```python
import cv2

# 영상 무한재생 프로그램

cap = cv2.VideoCapture("./movie/Game.mp4")

fps = cap.get(cv2.CAP_PROP_FPS) # 1초 동안 진행되는 프레임 = 프레임 속도
delay = (1.0/fps) * 1000

if cap.isOpened():
    count = 0
    while True:
        movie, frame = cap.read()
        if movie: # 영상을 받아오는 경우
            cv2.imshow("Game movie", frame)
            if cv2.waitKey(int(delay)) == 27:
                break
            count += 1
            print("count = ", count, "영상의 프레임 수 = ", cap.get(cv2.CAP_PROP_FRAME_COUNT))
        else:
            cap.set(cv2.CAP_PROP_POS_FRAMES, 0) # 동영상의 시작 부분으로 돌아감
cap.release()
cv2.destroyAllWindows()
```
