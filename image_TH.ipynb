```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("./photo/goose.jpg", cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (600, 600))

ret , dst = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)

cv2.imshow("dst", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
ret, dst = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
ret, dst_inv = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)

plt.subplot(1, 3, 1) #  
plt.imshow(img, cmap='gray')
plt.subplot(1, 3, 2)
plt.imshow(dst, cmap='gray')
plt.subplot(1, 3, 3)
plt.imshow(dst_inv, cmap='gray')
plt.show()
```

```python
ret, dst = cv2.threshold(img, 100, 255, cv2.THRESH_TRUNC)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.show()
```

```python
#  ->   

ret, dst = cv2.threshold(img, 100, 255, cv2.THRESH_MASK)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.show()
```

```python
#     

ret, dst = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

plt.subplot(1, 2, 1) #   
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.show()
```
