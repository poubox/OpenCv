```python
import cv2
cv2.namedWindow("My_Love", cv2.WINDOW_NORMAL)
img = cv2.imread("./photo/latte.png", cv2.IMREAD_ANYCOLOR)
# 이미지가 제대로 로드되었는지 확인
if img is None:
    print("Failed to load image.")
else:
    # 이미지를 창에 표시
    cv2.imshow("My_Love", img)

    # 키 입력 대기
    cv2.waitKey(0)

    # 창 닫기
    cv2.destroyWindow("My_Love")

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

cv2.namedWindow("My_Love", cv2.WINDOW_NORMAL)
img = cv2.imread("./photo/squares.png", cv2.IMREAD_ANYCOLOR)
cv2.imshow("My_Love", img)
cv2.waitKey(0)

if cv2.waitKey(10) == 9: # tab
    cv2.moveWindow("My_Love", 100, 100)
if cv2.waitKey(10) == 27: # esc
    cv2.destroyWindow("My_Love")

```

```python
img = cv2.imread("./photo/squares.png")
img_cvt = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
```
